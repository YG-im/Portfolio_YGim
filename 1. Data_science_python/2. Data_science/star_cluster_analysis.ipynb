## Radial coordinate 도입 및 $r~\rm{vs}~v_r$ 그래프

```python
df1['r'] = (df1['x']**2 + df1['y']**2 + df1['z']**2)**(1/2)
```

```python
df1['vr'] = (df1['vx']**2 + df1['vy']**2 + df1['vz']**2)**(1/2)
```

```python
df1['K'] = (1/2)*df1['m']*df1['vr']**2
```
